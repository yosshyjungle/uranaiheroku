```python
base_url = f'https://sp.gettersiida.net/article/gettersiida/unsei/{params}'
base_url
```

```python
req = requests.get(base_url)
soup = BeautifulSoup(req.text, 'html.parser', from_encoding='utf_8')
soup
```

```python
#class entry-contentの中身を取得
entry_contents = soup.find('div', attrs={'class':'entry-content'})
entry_contents
```
